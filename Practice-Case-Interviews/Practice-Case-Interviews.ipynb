### Life Sciences Industries
Regarding the life sciences industry, this means healthcare and related businesses like pharmaceuticals, medical equipment, biotechnology.


#### Most influential companies in the Netherlands


- Pharmaceuticals: **Abbott Labs** *(diagnostics, med devices, nutrition)*, **AstraZeneca** *(oncology, cardiovascular, renal, and respiratory)*, **Novartis** *(oncology and neuroscience)*;

- Biotechnology: **Galapagos** *(drug and antibody R&D)*, **Argenx** *(autoimmune and hematological cancers, antibody therapy)* **Cergentis** *(genomics)*, **Amgen** *(human therapeutics, biologics)*;

- Medical Equiment: **DutchMed Group** *(med equipment)*, **Van Straten Med** *(surgical, disposables)*, **Pie Med Imaging** *(cardiac imaging)*;

- Cosmeceuticals: **L'Oreal**, **Unilever**, **Special Cosmetics**, **HH Cosmetics**;

- Managed Healthcare (Insurances): **MSD** *(prescription, vaccines, bio therapy)*, **Johnson&Johnson** *(pharmaceuticals, medical devices)*

#### Key Trends to consider

**Big data**: Digitalization in the life sciences industry has resulted in the generation of huge volumes of data, which have to be stored and analyzed accurately to realize their full potential.

**Automation and modernization**: The majority of big pharma companies have been established through different mergers and acquisitions  like IQVIA and have adopted portfolios of IT applications in different phases of modernization. Theres a huge need for standardizing business processes, measuring manufacturing, focusing on visibility, and leveraging the right software to drive business value.

**Artificial intelligence**: AI-driven tools is now leveraged at all phases of drug research and development. For example AI-driven software can *assist in planning possible synthesis pathways in chemicals* to obtain compounds of interest. Innovations such as nanosensors and AI-powered imaging technologies are revolutionizing diagnostics, enabling earlier detection of diseases and improving treatment planning through detailed imaging analysis

**Revolutionizing supply chains**: Extensive remodeling of supply chains for strategic products, such as drug, medical equipment, diagnostics, food, and chemicals led by the national regulatory activities in response to the COVID-19 crisis

**Blockchain and wearables**: Wearable technology is altering the world and has paved the way for many people to monitor health, including sleeping patterns and overall fitness. Blockchain has helped wearable devices and mobile applications connect to a patient hub with all health records, allowing doctors to access a new level of visibility in real-time.

**Precision Medicine**: becoming increasingly prominent, allowing for tailored treatment plans based on individual genetic profiles, environments, and lifestyles. This approach contrasts with traditional "one-size-fits-all" methods and is particularly impactful in oncology, where targeted therapies are developed based on specific tumor characteristics


**Outsourcing**: Pharmaceutical companies are starting to outsource research programs to academic and private contract research organizations (CROs) to stay ahead of the competition in the world of rapidly growing knowledge and advanced technologies. Activities such as clinical trials, efficacy tests in animal models, and assay development.

**Telemedicine**: Progress in telemedicine is one of the biggest sources of substantial change in the healthcare ecosystem. It is improving the quality of diagnosis and treatment by allowing patients to get proper access to healthcare professionals.

### Step-by-Step to tackle the Case

1. Understand the case
    Handing over the Case (2 min): 
    - Take meticulous notes on the most important pieces of information. Focus on understanding the context of the situation and the **objective** of the case. 
    - Ask clarifying questions, **summarize the case** background information back to the interviewer to confirm my understanding of the case.

 The most important part of this step is to **verify the objective of the case**. Not answering the right business question is the quickest way to fail.

 (Solving the case: 45 mins)

2. Structure the problem
    - Break down the problem into smaller, more manageable components. Brainstorm different ideas and organizing them into different categories.
    - Id the major issues or areas that need to be explored, walk the interviewer through the framework
 
 The most important part is to outline how the case can be solved, without going into the details yet, making sure the big picture makes sense and tackles the **objective**

3. Kick off the case
    - This can be led by me or the interviewer, choose an area and give a reason why start here.

4. Quantitative problems
    - May be required to calculate a certain profitability or financial metric. Or be asked to estimate the size of a particular market or to estimate a particular figure.
    - First layout how such numbers could be calculated with some estimations and why they make sense, then check with the interviewer if it makes sense or missing any point, then just quick maths

5. Qualitative problems
    - May also have qualitative problems to answer, in this case brainstorm ideas based on key trends at the moment or business situations. Always clearly explain reasoning behind the summary of a certain stance or decision

(Presenting the case: 30 mins) 
 
6. Deliver a recommendation
    - Always finish by presenting a recommendations which acts as a summary of the case, presenting the potential solution to the **objective** and major reasons that support it.

(Ask questions about IQVIA or the role: 20 mins)

### Common Frameworks
Practice by writing down on paper, no calculators allowed so also practice math as well

#### Main Components of Frameworks

market attractiveness, customer needs, competitive landscape, company capabilities, profitability, risks, and strategic alternatives.

Its important to order these components logically in a framework, in the way that answering one enables the next

**Profitability**

        - Expected Revenue?
        - Expected Costs?
        - Time to break even?

        Revenue
            - Product Price
            - Quantity Sold

        Costs
            - Fixed Costs
            - Variable Costs


**Market Entry**

*Market Attractiveness*

        - What is the market size?
        - What is the market growth rate?
        - What are the contribution margins on average?

*Competitive Landscape*

        - How many competitors are there?
        - How much share does each have?
        - Are there competitive advantages like IP or differentiation factors?
        - Are there barriers to entry in the market?

*Company capabilities*

        - Does the company have the expertise and products required
        - Does the company have marketing/distribution channels required

*Customer Needs*

        - What are the customers preferences or needs?
        - What are their purchase habits/power?
        - What is the customers opinion on the company?

*Market Size*

        - Top Down Approach: Start with the maximum market available, then cut percentages of it until reaching the market of interest making reasonable estimations
        - Bottom Up Approach: Start with a small local statistic and extrapolate to the market of interest with reasonable estimations

*Risks*

        - What are major risks involved in this market?
        - What can be done to mitigate these risks?


**Merger or Acquisition**

    Company Attractiveness  
        - Is the target company profitable?
        - How much market share do they have?
        - Are they growing?
        - Do they have differentiatior or competitive advantages like IP
    
    Synergies: benefits from using one company instead of two
        - Cost benefits: Reduced headcount 
        - Revenue benefits: New benefits on product offering or price

    Financial Considerations:
        - Is the acquisition price attractive?
        - How much does the client expect to improve and grow the company target?
        - Expected Return on Investment?
        - Qualitative: how is the marketing or r&d affected

**Pricing**

    A max price --> Economic value of the product

    A competitive price --> price of the competition for similar products

    A min price --> production cost of the product


### Framework Exercises

#### --- Profitability

**Brainstorm**

    Analyse Current profitability - is it a revenue issue, costs issue or both?
    Are there changes in customer needs?
    Is there a significant change in competitive landscape, markets or overall market?
    Is there a problem in the company capabilities?

**Framework**

    Profitability issue  
        - Are the sales going down?  
        - Are the prices going down?
        - Have fixed costs gone up?
        - Have variable costs gone up?

    Customer needs
        - Have the customers preferences or needs changed?
        - Have their purchase habits/power changed?
        - Has the customers opinion on the company changed?

    Competitive landscape
        - Are there new competitors or competitors mergers/acquisitions?
        - Have leading competitors made recent big strategic changes?
        - Are there new technologies impacting the market?
        - Are there new regulations impacting the market?
    
    Company capabilities
        - Any recent changes in the company expertise? Talent leaving?
        - Are there recent strategic changes or leadership changes?
        - Are there management changes impacting production?


#### --- Competition

**Brainstorm**

    How threatening is the new casino?
    How will custumers and competition react to the new casino
    With this info consider strategic options and offer a recommendation

**Framework**

    New casino threat/capabilities
        - How large is the new casino, how many customers do they serve?
        - How close is the new casino to the clients casino?
        - Is there a differentiator or competitive advantage, like pricing strategy or product offering?

    Customer response
        - How will custumers react to the new casino?
        - How likely are they to switch to the new casino?

    Competition response
        - How will competitors react?
        - Open a new casino, or close nearby casino?
        - What else could competitors do?
    
    Strategic options
        - Do nothing
        - Open a new casino
        - Close the current casino
        - Improve attractiveness of current casino
        - Alternative?
